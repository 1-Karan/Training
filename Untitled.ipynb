```python
import numpy as np
import pandas as pd



info={
    "name": ["abhay","girish","priyanka","girishEx"],
    "age": [20,23,27,29],
    "course":["aiml", "cse", "blockchain", "cybersecurity"],
    "marks": [32,81,91,100]
}

dp=pd.DataFrame(info)
print(dp)
```

```python
print(dp["name"])
print(dp[["name","course"]])
print(dp.iloc[0])
print(dp.loc[2, "marks"])
```

```python

high_scorers=dp[dp["marks"] > 80]
print("high scorers: ",high_scorers)

dp["res"]=np.where(dp["marks"] >=80, "pass","fail")

dp.loc[dp["name"]=="priyanka", "marks"]= 95
print(dp)
```

```python
import matplotlib.pyplot as plt
avg_marks= dp.groupby("course")["marks"].mean()

avg_marks.plot(kind="bar",color="green", edgecolor="yellow")
plt.title("Avg Marks of Students")
plt.xlabel("course")
plt.ylabel("average marks")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Product": ["Laptop", "Laptop", "Laptop", "Mobile", "Mobile", "Mobile"],
    "Units_Sold": [120, 135, 150, 200, 220, 210],
    "Revenue": [600000, 675000, 750000, 300000, 330000, 315000]
}
 
df = pd.DataFrame(data)
avg_revenue= df.groupby("Product")["Revenue"].mean()

avg_revenue.plot(kind="bar",color="green", edgecolor="yellow")
plt.title("Comparison")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```
