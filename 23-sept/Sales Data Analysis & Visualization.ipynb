```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
total_revenue = df.groupby("Product")["Revenue"].sum()
print("Total Revenue per Product : ",total_revenue)
```

```python
avg_units_sold = df.groupby("Product")["Units_Sold"].mean()
print("Acerage UniTS SOLD per Product: ",avg_units_sold)
```

```python
plt.figure(figsize=(8,5))
for product in df["Product"].unique():
    product_data = df[df["Product"] == product]
    plt.plot(product_data["Month"],product_data["Revenue"], marker='o', label=product)

    plt.title("Monthly Revenue Trend (Jan-Mar)")
    plt.xlabel("Month")
    plt.ylabel("Revenue")
    plt.legend()
    plt.show()
```

```python
total_revenue.plot(kind="bar",color=["Blue","Green","Orange"])
plt.title("Contribution of each product in Revenue")
plt.ylabel("")
plt.show()
                   
```

```python
total_revenue.plot(kind="pie")
plt.title("Contribution of each product to Total Revenue: ")
plt.ylabel("")
plt.show()
```
